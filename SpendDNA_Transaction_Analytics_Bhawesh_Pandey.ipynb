# Feature 1 : Transaction Parser

The first step in any data analysis project is to clean the raw dataset.

In this feature, we will:

- Parse different date formats
- Clean amount values
- Standardize transaction types
- Remove duplicate transactions
- Prepare a clean dataset for further analysis

## Step 1 : Parse Date Column

### Objective

Convert all date values into a standard datetime format so Python can understand and analyze dates correctly.



```python

# Feature 1 : Transaction Parser
# Step 1 : Parse Date Column


df["Date"] = pd.to_datetime(
    df["Date"],
    format="mixed",
    dayfirst=True,
    errors="coerce"
)

print("=" * 60)
print("DATE PARSING SUMMARY")
print("=" * 60)

print("\nData Type:")
print(df["Date"].dtype)

print("\nUnparseable Dates:")
print(df["Date"].isna().sum())
```

## Step 2 : Clean Amount Column

### Objective

Convert the Amount column into a numeric format by removing currency symbols, commas, and unnecessary spaces.



```python

# Feature 1 : Transaction Parser
# Step 2 : Clean Amount Column


# Remove currency symbols, commas and extra spaces
df["Amount"] = (
    df["Amount"]
    .astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace("Rs.", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.strip()
)

# Convert Amount column into numeric datatype
df["Amount"] = pd.to_numeric(
    df["Amount"],
    errors="coerce"
)

print("=" * 60)
print("AMOUNT CLEANING SUMMARY")
print("=" * 60)

print("\nData Type:")
print(df["Amount"].dtype)

print("\nUnparseable Amounts:")
print(df["Amount"].isna().sum())
```

## Step 3 : Standardize Transaction Type

### Objective

Convert all transaction types into a consistent format.



We will convert:

- Debit → debit
- DR → debit
- CR → credit

This ensures consistency throughout the project.

```python

# Feature 1 : Transaction Parser
# Step 3 : Standardize Transaction Type


# Display transaction types before cleaning
print("=" * 60)
print("TRANSACTION TYPES BEFORE STANDARDIZATION")
print("=" * 60)

print(df["Type"].unique())

# Convert all values to lowercase
df["Type"] = df["Type"].str.lower()

# Standardize transaction types
df["Type"] = df["Type"].replace({
    "dr": "debit",
    "cr": "credit"
})

print("\n" + "=" * 60)
print("TRANSACTION TYPES AFTER STANDARDIZATION")
print("=" * 60)

print(df["Type"].unique())
```

## Step 4 : Remove Duplicate Transactions

### Objective

Identify and remove duplicate transaction records from the dataset.



```python

# Feature 1 : Transaction Parser
# Step 4 : Remove Duplicate Transactions


# Count duplicate rows before removing
duplicates_before = df.duplicated().sum()

print("=" * 60)
print("DUPLICATE TRANSACTION SUMMARY")
print("=" * 60)

print(f"\nDuplicate Rows Before Removal : {duplicates_before}")

# Remove duplicate rows
df = df.drop_duplicates()

# Count duplicate rows after removing
duplicates_after = df.duplicated().sum()

print(f"Duplicate Rows After Removal  : {duplicates_after}")

print(f"\nFinal Dataset Shape : {df.shape}")
```

### Observation

The Transaction Parser successfully cleaned the raw transaction dataset by performing the following operations:

- Parsed all mixed date formats into a standard datetime format.
- Converted transaction amounts into numeric values.
- Standardized transaction types into a consistent format (debit and credit).
- Removed all duplicate transaction records.

After preprocessing, the dataset contains **1,310 unique transactions** and is now ready for further analysis.

## Step 4 : Build Vendor Extraction Function

### Objective

Create a function to identify and standardize vendor names from transaction descriptions.

### Why?

The same merchant can appear in different formats (UPI, POS, BHIM, IMPS, NEFT, etc.). Standardizing these descriptions into a single vendor name makes spending analysis accurate and avoids duplicate vendor entries.

```python

# Feature 2 : Vendor Extractor
# Step 4 : Build Vendor Extraction Function


def extract_vendor(description):

    description = str(description).upper().strip()

    # -------------------------------------------------------
    # Salary & Income
    # -------------------------------------------------------
    if "SALARY" in description:
        return "Salary"

    # -------------------------------------------------------
    # ATM Withdrawals
    # -------------------------------------------------------
    if "ATM-WDL" in description:
        return "ATM Withdrawal"

    # -------------------------------------------------------
    # Rent
    # -------------------------------------------------------
    if "RENT" in description:
        return "Rent"

    # -------------------------------------------------------
    # Investments
    # -------------------------------------------------------
    if "ZERODHA" in description:
        return "Zerodha"

    if "GROWW" in description:
        return "Groww"

    # -------------------------------------------------------
    # Quick Commerce
    # -------------------------------------------------------
    if "INSTAMART" in description:
        return "Swiggy Instamart"

    if "BLINKIT" in description or "GROFERS" in description:
        return "Blinkit"

    if "ZEPTO" in description or "KIRANAKART" in description:
        return "Zepto"

    if "BIGBASKET" in description:
        return "BigBasket"

    if "DMART" in description or "AVENUE SUPERMARTS" in description:
        return "DMart"

    # -------------------------------------------------------
    # Food Delivery
    # -------------------------------------------------------
    if "SWIGGY" in description or "BUNDL TECH" in description:
        return "Swiggy"

    if "ZOMATO" in description:
        return "Zomato"

    # -------------------------------------------------------
    # Shopping
    # -------------------------------------------------------
    if "AMAZON" in description or "AMZN" in description:
        return "Amazon"

    if "FLIPKART" in description or "FKART" in description:
        return "Flipkart"

    if "MYNTRA" in description:
        return "Myntra"

    if "NYKAA" in description or "FSN E-COMMERCE" in description:
        return "Nykaa"

    # -------------------------------------------------------
    # Transport
    # -------------------------------------------------------
    if "UBER" in description:
        return "Uber"

    if "ANI TECHNOLOGIES" in description or "OLA" in description:
        return "Ola"

    if "RAPIDO" in description:
        return "Rapido"

    if "BMTC" in description or "TUMMOC" in description:
        return "BMTC"

    # -------------------------------------------------------
    # Entertainment
    # -------------------------------------------------------
    if (
        "BOOKMYSHOW" in description
        or "BIGTREE" in description
        or "BMS MOVIE" in description
    ):
        return "BookMyShow"

    if "NETFLIX" in description:
        return "Netflix"

    if "SPOTIFY" in description:
        return "Spotify"

    if (
        "HOTSTAR" in description
        or "STAR INDIA" in description
        or "DISNEY" in description
    ):
        return "Disney+ Hotstar"

    # -------------------------------------------------------
    # Telecom
    # -------------------------------------------------------
    if "AIRTEL" in description:
        return "Airtel"

    if "JIO" in description:
        return "Jio"

    if "VI" in description or "VODAFONE" in description:
        return "Vi"

    # -------------------------------------------------------
    # Utilities
    # -------------------------------------------------------
    if "BESCOM" in description or "BANGALORE ELEC" in description:
        return "BESCOM"

    if "BWSSB" in description:
        return "BWSSB"

    # -------------------------------------------------------
    # Fuel
    # -------------------------------------------------------
    if "INDIAN OIL" in description or "IOC" in description:
        return "Indian Oil"

    if "BPCL" in description:
        return "BPCL"

    if "HP PETROL" in description:
        return "HP Petrol"

    # -------------------------------------------------------
    # Coffee & Cafes
    # -------------------------------------------------------
    if "STARBUCKS" in description or "TATA STARBUCKS" in description:
        return "Starbucks"

    if "COFFEE DAY" in description or "CCD" in description:
        return "Cafe Coffee Day"

    if (
    "THIRD WAVE" in description
    or "THIRDWAVE" in description
    or "TWC INDIA" in description
):
      return "Third Wave Coffee"

    if "TRUFFLES" in description:
        return "Truffles"

    if "MEGHANA" in description:
        return "Meghana Foods"

    if "EMPIRE" in description:
        return "Empire Restaurant"

  # Restaurants
    if "DINEOUT" in description:
        return "Dineout"

    if "RESTAURANT" in description:
      return "Restaurant"

# Public Transport
    if "ROPPEN" in description:
      return "BMTC"


    # -------------------------------------------------------
    # Personal Transfers
    # -------------------------------------------------------
    if any(name in description for name in [
        "AMAN", "ANKIT", "PRIYA",
        "NEHA", "SNEHA", "VIKAS",
        "KARAN"
    ]):
        return "Personal Transfer"

    # -------------------------------------------------------
    # Default
    # -------------------------------------------------------
    return "Others"
```

# Feature 3 : Category Tagger

The Category Tagger classifies every transaction into a meaningful spending category based on its extracted vendor.

Instead of analyzing hundreds of individual vendors, categorizing transactions makes it easier to understand spending habits, generate summaries, and create insightful visualizations.

## Step 1 : Build Category Tagging Function

### Objective

Create a function that assigns a spending category to each transaction based on its extracted vendor.


```python
def assign_category(vendor):

    # -----------------------------
    # Food Delivery
    # -----------------------------
    if vendor in [
        "Swiggy",
        "Zomato"
    ]:
        return "Food Delivery"

    # -----------------------------
    # Restaurants
    # -----------------------------
    elif vendor in [
        "Restaurant",
        "Empire Restaurant",
        "Meghana Foods",
        "Truffles",
        "Dineout"
    ]:
        return "Restaurants"

    # -----------------------------
    # Cafe
    # -----------------------------
    elif vendor in [
        "Starbucks",
        "Cafe Coffee Day",
        "Third Wave Coffee"
    ]:
        return "Cafe"

    # -----------------------------
    # Quick Commerce
    # -----------------------------
    elif vendor in [
        "Blinkit",
        "Zepto",
        "Swiggy Instamart"
    ]:
        return "Quick Commerce"

    # -----------------------------
    # Groceries
    # -----------------------------
    elif vendor in [
        "BigBasket",
        "DMart"
    ]:
        return "Groceries"

    # -----------------------------
    # E-Commerce
    # -----------------------------
    elif vendor in [
        "Amazon",
        "Flipkart",
        "Myntra",
        "Nykaa"
    ]:
        return "E-Commerce"

    # -----------------------------
    # Transport
    # -----------------------------
    elif vendor in [
        "Uber",
        "Ola",
        "Rapido",
        "BMTC"
    ]:
        return "Transport"

    # -----------------------------
    # Fuel
    # -----------------------------
    elif vendor in [
        "Indian Oil",
        "BPCL",
        "HP Petrol"
    ]:
        return "Fuel"

    # -----------------------------
    # Entertainment
    # -----------------------------
    elif vendor in [
        "Netflix",
        "Spotify",
        "Disney+ Hotstar",
        "BookMyShow"
    ]:
        return "Entertainment"

    # -----------------------------
    # Telecom
    # -----------------------------
    elif vendor in [
        "Airtel",
        "Jio",
        "Vi"
    ]:
        return "Telecom"

    # -----------------------------
    # Utilities
    # -----------------------------
    elif vendor in [
        "BESCOM",
        "BWSSB"
    ]:
        return "Utilities"

    # -----------------------------
    # Investments
    # -----------------------------
    elif vendor in [
        "Groww",
        "Zerodha"
    ]:
        return "Investments"

    # -----------------------------
    # Income
    # -----------------------------
    elif vendor == "Salary":
        return "Income"

    # -----------------------------
    # Housing
    # -----------------------------
    elif vendor == "Rent":
        return "Housing"

    # -----------------------------
    # Cash Withdrawal
    # -----------------------------
    elif vendor == "ATM Withdrawal":
        return "Cash Withdrawal"

    # -----------------------------
    # Personal Transfer
    # -----------------------------
    elif vendor == "Personal Transfer":
        return "Personal Transfer"

    else:
        return "Others"
```

# Feature 4 : Spending Overview

The Spending Overview provides a high-level summary of the user's financial transactions.

It analyzes the total income, total expenses, transaction count, and net savings, giving a quick snapshot of overall financial activity. This summary serves as the foundation for all subsequent spending analyses and visualizations.

Calculate the key financial metrics from the transaction dataset.


```python

# Feature 4 : Spending Overview
# Step 1 : Generate Overall Spending Summary


total_transactions = len(df)

total_income = df.loc[df["Type"] == "credit", "Amount"].sum()

total_expense = df.loc[df["Type"] == "debit", "Amount"].sum()

net_savings = total_income - total_expense

print(f"Total Transactions : {total_transactions}")
print(f"Total Income       : ₹{total_income:,.2f}")
print(f"Total Expense      : ₹{total_expense:,.2f}")
print(f"Net Savings        : ₹{net_savings:,.2f}")
```

### Objective

Present the calculated financial metrics in a structured summary table.


```python
summary = pd.DataFrame({
    "Metric": [
        "Total Transactions",
        "Total Income",
        "Total Expense",
        "Net Savings"
    ],
    "Value": [
        total_transactions,
        f"₹{total_income:,.2f}",
        f"₹{total_expense:,.2f}",
        f"₹{net_savings:,.2f}"
    ]
})

summary
```

# Feature 5 : Monthly Spending Trends

The Monthly Spending Trends feature analyzes how income and expenses change over time.

It summarizes monthly financial activity, identifies spending patterns, and highlights trends that can support better financial planning and decision-making.

## Step 1 : Extract Month and Year

### Objective

Extract the month and year from the transaction date.



```python

# Feature 5 : Monthly Spending Trends
# Step 1 : Extract Month and Year


df["Month"] = df["Date"].dt.strftime("%b")

df["Year"] = df["Date"].dt.year

df["Month_Year"] = df["Date"].dt.strftime("%b-%Y")

df[["Date", "Month", "Year", "Month_Year"]].head()
```

## Step 2 : Monthly Income and Expense Summary

### Objective

Calculate monthly income and monthly expenses.



```python
monthly_summary = (
    df.groupby(["Month_Year", "Type"])
      .agg(
          Total_Amount=("Amount", "sum")
      )
      .reset_index()
)

monthly_summary
```

## Step 3 : Create Monthly Summary Table

### Objective

Convert the grouped monthly data into a pivot table.



```python
monthly_summary = (
    monthly_summary
    .pivot(
        index="Month_Year",
        columns="Type",
        values="Total_Amount"
    )
    .fillna(0)
    .reset_index()
)

monthly_summary
```

## Step 4 : Sort Monthly Summary Chronologically

### Objective

Arrange the monthly summary in chronological order.



```python
monthly_summary = (
    df.groupby([
        pd.Grouper(key="Date", freq="MS"),
        "Type"
    ])
    .agg(
        Total_Amount=("Amount", "sum")
    )
    .reset_index()
)

monthly_summary["Month_Year"] = monthly_summary["Date"].dt.strftime("%b-%Y")

monthly_summary = (
    monthly_summary
    .pivot(
        index=["Date", "Month_Year"],
        columns="Type",
        values="Total_Amount"
    )
    .fillna(0)
    .reset_index()
)

monthly_summary
```

## Step 5 : Create Category-wise Monthly Spending Matrix

### Objective

Create a month-wise spending matrix for every spending category.



```python

# Feature 5 : Monthly Spending Trends
# Step 5 : Category-wise Monthly Spending Matrix


category_monthly = (
    df[df["Type"] == "debit"]
    .groupby(["Category", "Month_Year"])
    .agg(
        Total_Spending=("Amount", "sum")
    )
    .reset_index()
)

category_monthly
```

## Step 6 : Convert to Matrix Format

### Objective

Convert the grouped monthly spending data into a matrix.


```python
category_monthly_matrix = (
    category_monthly
    .pivot(
        index="Category",
        columns="Month_Year",
        values="Total_Spending"
    )
    .fillna(0)
)
```

```python
month_order = [
    "Jan-2024",
    "Feb-2024",
    "Mar-2024",
    "Apr-2024",
    "May-2024",
    "Jun-2024"
]

category_monthly_matrix = category_monthly_matrix.reindex(
    columns=month_order
)

category_monthly_matrix
```

## Step 7 : Identify Highest Spending Category Each Month

### Objective

Determine which spending category recorded the highest expenditure in each month.



```python
highest_category = category_monthly_matrix.idxmax()

highest_amount = category_monthly_matrix.max()

monthly_highest = pd.DataFrame({
    "Highest Category": highest_category,
    "Amount": highest_amount
})

monthly_highest
```

## Step 8 : Monthly Trend Report

### Objective

Generate a text-based monthly spending report.



```python
print("=" * 60)
print("MONTHLY SPENDING TREND REPORT")
print("=" * 60)

for month in month_order:
    if month in monthly_highest.index:
        category = monthly_highest.loc[month, "Highest Category"]
        amount = monthly_highest.loc[month, "Amount"]

        print(f"{month:<10} | {category:<20} | ₹{amount:,.2f}")
```

## Step 9 : Analyze Monthly Spending Growth

### Objective

Calculate the month-to-month change in total monthly expenses.



```python
monthly_growth = monthly_summary.copy()

monthly_growth["Monthly_Change"] = (
    monthly_growth["debit"].diff()
)

monthly_growth["Growth (%)"] = (
    monthly_growth["debit"]
    .pct_change()
    * 100
).round(2)

monthly_growth
```

```python
print("=" * 70)
print("MONTHLY GROWTH REPORT")
print("=" * 70)

for _, row in monthly_growth.iterrows():

    if pd.isna(row["Monthly_Change"]):
        change = "N/A"
    elif row["Monthly_Change"] >= 0:
        change = f"+₹{row['Monthly_Change']:,.0f}"
    else:
        change = f"-₹{abs(row['Monthly_Change']):,.0f}"

    print(
        f"{row['Month_Year']:<10}"
        f" Expense: ₹{row['debit']:>9,.0f}"
        f"   Change: {change}"
    )
```

### Observation

The Monthly Spending Trends feature successfully analyzed spending patterns across different months.

Key outcomes include:

- Generated month-wise income and expense summaries.
- Created a category-wise monthly spending matrix.
- Identified the highest spending category for each month.
- Calculated month-over-month changes in expenditure to highlight financial trends.
- Produced a text-based monthly trend report in accordance with the project constraints.

# Feature 6 : Time-of-Day Analysis

The Time-of-Day Analysis examines when financial transactions occur throughout the day.

By grouping transactions into different time periods, this feature helps identify spending habits based on the time of transaction.

## Step 1 : Extract Transaction Hour

### Objective

Extract the transaction hour from the timestamp.



```python

# Feature 6 : Time-of-Day Analysis
# Step 1 : Extract Transaction Hour


df["Hour"] = pd.to_datetime(
    df["Time"],
    format="%H:%M"
).dt.hour

df[["Time", "Hour"]].head(10)
```

## Step 2 : Classify Time of Day

### Objective

Group every transaction into a time-of-day category.



```python
def time_period(hour):

    if 5 <= hour < 12:
        return "Morning"

    elif 12 <= hour < 17:
        return "Afternoon"

    elif 17 <= hour < 21:
        return "Evening"

    else:
        return "Night"


df["Time_Period"] = df["Hour"].apply(time_period)

df[["Time", "Hour", "Time_Period"]].head(15)
```

## Step 3 : Analyze Transactions by Time Period

### Objective

Calculate the number of transactions and total spending for each time period.



```python
time_summary = (
    df[df["Type"] == "debit"]
    .groupby("Time_Period")
    .agg(
        Transactions=("Amount", "count"),
        Total_Spending=("Amount", "sum")
    )
)

time_summary
```

## Step 4 : Sort Time Periods

### Objective

Arrange the time periods in chronological order.

# Feature 7 : Anomaly Detection

The Anomaly Detection feature identifies unusually large transactions within each spending category using the Z-score statistical method.

Instead of comparing every transaction globally, each transaction is compared only with other transactions belonging to the same category.

Transactions having a Z-score greater than 2 are considered anomalous.

## Step 1 : Calculate Category-wise Mean and Standard Deviation

### Objective

Calculate the average spending and standard deviation for every category.



```python

# Feature 7
# Step 1 : Category Statistics

debit_df = df[df["Type"] == "debit"].copy()

debit_df["Category_Mean"] = (
    debit_df.groupby("Category")["Amount"]
    .transform("mean")
)

debit_df["Category_STD"] = (
    debit_df.groupby("Category")["Amount"]
    .transform("std")
)

debit_df.head()
```

## Step 2 : Calculate Z-Score

### Objective

Calculate the Z-score for every debit transaction.



```python
debit_df["Z_Score"] = (
    (debit_df["Amount"] - debit_df["Category_Mean"])
    / debit_df["Category_STD"]
)

debit_df.head()
```

## Step 3 : Detect Anomalous Transactions

### Objective

Identify unusually large transactions.



```python
anomalies = debit_df[
    debit_df["Z_Score"] > 2
].copy()

anomalies = anomalies.sort_values(
    by="Z_Score",
    ascending=False
)

print("Total Anomalies :", len(anomalies))

anomalies[
    [
        "Date",
        "Vendor",
        "Category",
        "Amount",
        "Z_Score"
    ]
].head(10)
```

## Step 4 : Category-wise Anomaly Summary

### Objective

Summarize anomalies across spending categories.



```python
category_anomalies = (
    anomalies.groupby("Category")
    .agg(
        Total_Anomalies=("Amount", "count"),
        Total_Anomaly_Amount=("Amount", "sum")
    )
    .sort_values(
        by="Total_Anomalies",
        ascending=False
    )
)

category_anomalies
```

## Step 5 : Anomaly Detection Report

### Objective

Generate a concise anomaly summary.



```python
print("=" * 70)
print("ANOMALY DETECTION REPORT")
print("=" * 70)

print(f"Total Debit Transactions : {len(debit_df)}")
print(f"Total Anomalies          : {len(anomalies)}")

print("\nTop 5 Anomalies\n")

for _, row in anomalies.head(5).iterrows():
    print(
        f"{row['Date'].date()} | "
        f"{row['Vendor']:<20} | "
        f"{row['Category']:<18} | "
        f"₹{row['Amount']:>10,.2f} | "
        f"Z = {row['Z_Score']:.2f}"
    )
```

### Observation

The Z-score method successfully identified transactions that were significantly higher than the average spending within their respective categories.

Key outcomes include:

- Computed category-wise mean and standard deviation.
- Calculated the Z-score for every debit transaction.
- Flagged transactions with a Z-score greater than 2.
- Generated category-wise anomaly summaries and a formatted anomaly report.

This approach follows the project specification and provides category-specific anomaly detection.

# Feature 8 : Spending Archetype Detection

The Spending Archetype Detection feature analyzes Rahul's transaction history and assigns one or more financial personalities based on predefined spending rules.

Unlike traditional classification, a user can satisfy multiple archetypes simultaneously. Each detected archetype is supported with quantitative evidence from the spending analysis.

## Step 1 : Calculate Spending Metrics

### Objective

Calculate all metrics required for archetype detection.



```python

# Feature 8 : Fresh Metrics


debit_df = df[df["Type"] == "debit"].copy()
credit_df = df[df["Type"] == "credit"].copy()

# Totals
total_income = credit_df["Amount"].sum()
total_expense = debit_df["Amount"].sum()
net_savings = total_income - total_expense

if total_income > 0:
    savings_rate = (net_savings / total_income) * 100
else:
    savings_rate = 0

# Category spending
category_spending = (
    debit_df.groupby("Category")["Amount"]
    .sum()
)

category_percentage = (
    category_spending / total_expense * 100
)

food_percentage = (
    category_percentage.get("Food Delivery", 0)
    + category_percentage.get("Restaurants", 0)
    + category_percentage.get("Cafe", 0)
)

quick_percentage = category_percentage.get("Quick Commerce", 0)

shopping_percentage = category_percentage.get("E-Commerce", 0)

investment_percentage = category_percentage.get("Investments", 0)

transport_percentage = category_percentage.get("Transport", 0)

print("Total Income :", total_income)
print("Total Expense :", total_expense)
print("Savings Rate :", round(savings_rate,2))

print("\nCategory Percentages\n")
print(category_percentage.round(2))
```

## Step 2 : Define Archetype Detection Functions

### Objective

Create independent functions for each spending archetype.



```python

# Archetype Functions

def foodie():
    return food_percentage > 25


def quick_commerce_junkie():
    return quick_percentage > 15


def shopaholic():
    return shopping_percentage > 15


def investor():
    return investment_percentage > 15


def cab_commuter():
    return transport_percentage > 10


def yolo_spender():
    return savings_rate < 10


def disciplined_saver():
    return savings_rate > 40
```

## Step 3 : Late-Night Snacker Detection

### Objective

Identify whether the majority of Food Delivery transactions occur late at night.


```python
food_orders = debit_df[
    debit_df["Category"] == "Food Delivery"
]

late_night_orders = food_orders[
    (food_orders["Hour"] >= 21)
    |
    (food_orders["Hour"] <= 2)
]

if len(food_orders) > 0:
    late_night_percentage = (
        len(late_night_orders)
        /
        len(food_orders)
    ) * 100
else:
    late_night_percentage = 0
```

## Step 4 : Subscription Lover Detection

### Objective

Determine whether Rahul actively maintains multiple digital subscriptions.



```python
subscription_vendors = [
    "Netflix",
    "Spotify",
    "Disney+ Hotstar"
]

active_subscriptions = (
    debit_df[
        debit_df["Vendor"].isin(subscription_vendors)
    ]["Vendor"]
    .nunique()
)
```

```python

# Late Night Snacker


food_orders = debit_df[
    debit_df["Category"] == "Food Delivery"
]

late_night_orders = food_orders[
    (food_orders["Hour"] >= 21)
    |
    (food_orders["Hour"] <= 2)
]

if len(food_orders) > 0:
    late_night_percentage = (
        len(late_night_orders) / len(food_orders)
    ) * 100
else:
    late_night_percentage = 0


def late_night_snacker():
    return late_night_percentage > 50
```

### Objective

Evaluate every archetype rule and generate the final financial personality profile.


```python
archetypes = []

if foodie():
    archetypes.append(
        f"THE FOODIE ({food_percentage:.2f}% on food)"
    )

if quick_commerce_junkie():
    archetypes.append(
        f"THE QUICK COMMERCE JUNKIE ({quick_percentage:.2f}% on Quick Commerce)"
    )

if shopaholic():
    archetypes.append(
        f"THE SHOPAHOLIC ({shopping_percentage:.2f}% on E-Commerce)"
    )

if investor():
    archetypes.append(
        f"THE INVESTOR ({investment_percentage:.2f}% on Investments)"
    )

if late_night_snacker():
    archetypes.append(
        f"THE LATE-NIGHT SNACKER ({late_night_percentage:.2f}% late-night food orders)"
    )

if subscription_lover():
    archetypes.append(
        f"THE SUBSCRIPTION LOVER ({active_subscriptions} active subscriptions)"
    )

if cab_commuter():
    archetypes.append(
        f"THE CAB COMMUTER ({transport_percentage:.2f}% transport spending)"
    )

if yolo_spender():
   archetypes.append(
    f"THE YOLO SPENDER (Expenses exceed income by {abs(savings_rate):.2f}%)"
)

if disciplined_saver():
    archetypes.append(
        f"THE DISCIPLINED SAVER (Savings Rate = {savings_rate:.2f}%)"
    )
```
